```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from textblob import TextBlob

file_path = '/content/247500_totaloutput_9parts.csv'
data = pd.read_csv(file_path, nrows=2000)

data.head()


# Function to clean the text data
def clean_text(text):
    """
    Utility function to clean text by removing links, special characters
    using simple regex statements.
    """

    text = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text)
    return text


def get_tweet_sentiment(tweet):
    """
    Utility function to classify sentiment of passed tweet
    using textblob's sentiment method
    """

    analysis = TextBlob(clean_text(tweet))
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

data['clean_text'] = data['text'].apply(clean_text)
data['sentiment'] = data['clean_text'].apply(get_tweet_sentiment)

data[['text', 'clean_text', 'sentiment']].head()

positive_data = data[data['sentiment'] == 'positive']
negative_data = data[data['sentiment'] == 'negative']
neutral_data = data[data['sentiment'] == 'neutral']

positive_data['tweetcreated'] = pd.to_datetime(positive_data['tweetcreated'])
negative_data['tweetcreated'] = pd.to_datetime(negative_data['tweetcreated'])
neutral_data['tweetcreated'] = pd.to_datetime(neutral_data['tweetcreated'])

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 15))

# Positive Sentiment
axes[0].hist(positive_data['tweetcreated'], bins=50, color='green')
axes[0].set_title('Tweet Frequency - Positive Sentiment')
axes[0].set_xlabel('Date and Time')
axes[0].set_ylabel('Number of Tweets')

# Negative Sentiment
axes[1].hist(negative_data['tweetcreated'], bins=50, color='red')
axes[1].set_title('Tweet Frequency - Negative Sentiment')
axes[1].set_xlabel('Date and Time')
axes[1].set_ylabel('Number of Tweets')

# Neutral Sentiment
axes[2].hist(neutral_data['tweetcreated'], bins=50, color='blue')
axes[2].set_title('Tweet Frequency - Neutral Sentiment')
axes[2].set_xlabel('Date and Time')
axes[2].set_ylabel('Number of Tweets')

plt.tight_layout()
plt.show()


sentiment_counts = data['sentiment'].value_counts()

plt.figure(figsize=(8, 5))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Sentiment Analysis Histogram')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
